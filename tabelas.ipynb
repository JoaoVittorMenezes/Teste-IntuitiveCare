```python
import pdfplumber
import pandas as pd

# Caminho do PDF
pdf_path = "tables\Anexo_I_Rol_2021RN_465.2021_RN627L.2024.pdf"

# Criar um DataFrame vazio para armazenar os dados
df_final = pd.DataFrame()

with pdfplumber.open(pdf_path) as pdf:
    for page in pdf.pages:
        table = page.extract_table()
        if table:
            df = pd.DataFrame(table[1:], columns=table[0])  # Criar DataFrame da tabela
            df_final = pd.concat([df_final, df], ignore_index=True)  # Adicionar ao DF final

# Exibir os primeiros dados para conferência
df_final.head()
```

```python
df_final.head()
```

```python
df_final.rename(columns={"OD": "Seg. Odontológica"}, inplace=True)


df_final.rename(columns={"AMB": "Seg. Ambulatorial"}, inplace=True)

```

```python
df_final.to_csv("tabelas_extraidas.csv", index=False)
```
